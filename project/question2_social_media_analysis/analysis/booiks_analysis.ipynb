```python
global_mean = df["rating"].mean()
global_std = df["rating"].std()

rating_summary = df.groupby("category")["rating"].describe()
outlier_categories = rating_summary[
    (rating_summary["count"] < 5) | 
    (abs(rating_summary["mean"] - global_mean) > global_std) |
    (rating_summary["std"] > 2)  # very polarized
]
print(outlier_categories)
```
